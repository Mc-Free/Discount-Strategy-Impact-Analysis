## 1. Import a csv file to DataFrame



```python
url = "https://drive.google.com/file/d/1MHjM8PtzvqIqwSwnoCh_Y803txx6BfNt/view?usp=drive_link" # orderlines.csv
path = "https://drive.google.com/uc?export=download&id="+url.split("/")[-2]
orderlines_df = pd.read_csv(path)
```

# 5. Checking rows with the largest & smallest value

```python
#Largest value

orderlines_df.nlargest(5, "product_quantity")
```

```python
#Smallest value

orderlines_df.nsmallest(5, "product_quantity")
```

```python
#Check columns with product quantities > 100

orderlines_df.loc[orderlines_df["product_quantity"] > 100]

# orderlines_df.query("product_quantity > 100")
```

```python
# Importing orders.csv DataFrame

url = "https://drive.google.com/file/d/18c-Z5VBy2hcUlnjnvfpqAHFhzMR1e36e/view?usp=drive_link" # orders.csv
path = "https://drive.google.com/uc?export=download&id="+url.split("/")[-2]
orders_df = pd.read_csv(path)
```

```python
# merging orders_df and orderlines_df

orders_df.merge(orderlines_df, right_on="id_order", left_on="order_id").query("total_paid > 600")
```
